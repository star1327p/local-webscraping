```python
import pandas as pd
from bs4 import BeautifulSoup

# Original URL: https://nrp.web.irs.gov/1040-study.html#tab=tab1

# Open and read the HTML file
with open("NRP_1040_Study_Webpage.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, "html.parser")
```

```python
StudyYear = []
SampleSize = []
PercentComplete = []

# This HTML webpage has two tables, and we only want the first one.
tables = soup.find_all("table")
tab = tables[0]

for row in tab.find("tbody").find_all("tr"):
    col = row.find_all("td")
    if col != []:
        StudyYear.append(col[0].text)
        SampleSize.append(col[1].text)
        PercentComplete.append(col[2].text)

test_data = pd.DataFrame({"StudyYear":StudyYear, "SampleSize":SampleSize, "PercentComplete":PercentComplete})
test_data["StudyYear"] = test_data["StudyYear"].astype(int)
test_data["SampleSize"] = test_data["SampleSize"].str.replace(',', '').astype(int)
test_data
```
